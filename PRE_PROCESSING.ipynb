```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
folder_path = './data'

dataframes = []
```

```python
# Loop melalui setiap file dalam folder
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):  # Pastikan hanya membaca file CSV
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path)
        dataframes.append(df)

# Menggabungkan semua Dataframe menjadi satu Dataframe
df = pd.concat(dataframes, ignore_index=True)
```

```python
df
```

```python
#menyimpan hasil gabungan dataset pada folder baru
new_folder_path = 'new_data'
output_file = os.path.join(new_folder_path, 'combined_dataset.csv')
if os.path.exists(output_file):
    os.remove(output_file)
df.to_csv(output_file, index = False)
```

```python
plt.figure(figsize=(12,5))

ax=sns.barplot(y=df_fix.sort_values(by='Loves', ascending=False).head(10).Title, 
            x=df_fix.sort_values(by='Loves', ascending=False).head(10).Loves,
           palette='Set2')

for p in ax.patches:
    width = p.get_width()
    plt.text(18+p.get_width(), p.get_y()+0.55*p.get_height(),
             '{:1.0f}'.format(width),
             ha='center', va='center')

sns.set(style='darkgrid')

plt.xticks()

plt.xlabel('Loves Count')
plt.ylabel('Food Recipe')
plt.title('Top 10 Most Loved Food Recipe', fontweight="bold")
plt.show()
```
